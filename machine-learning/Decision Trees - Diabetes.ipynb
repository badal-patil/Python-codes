```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
col_names = ['Pregnant', 'Glucose', 'BP', 'Skin', 'Insulin', 'BMI', 'Pedigree', 'Age', 'Label']
diab = pd.read_csv("C:/Users/user/Desktop/Badal's Program/ML Datasets/diabetes.csv", header = None, names= col_names)
diab.head()
```

```python
diab.info
```

```python
feat_cols = ['Pregnant', 'Glucose', 'BP', 'Insulin', 'BMI', 'Pedigree', 'Age']
x = diab[feat_cols]
y = diab.Label
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
dt1 = DecisionTreeClassifier()
dt1.fit(x_train, y_train)

y_pred = dt1.predict(x_test)
```

```python
print("Accuracy  :  ", metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn import tree
tree.plot_tree(dt1)
```

```python
dt2 = DecisionTreeClassifier(max_depth = 3, random_state = 0)
dt2=dt2.fit(x_train, y_train)

y_pred = dt2.predict(x_test)
```
